```python
#importing the required libraries
import numpy as np
import pandas as pd

```

```python
#Task-1
#Creating a pandas dataframe with numpy random values (4 features and 4 observation)
df=pd.DataFrame([np.random.rand(4),np.random.rand(4),np.random.rand(4),np.random.rand(4)]) 
print(df)
```

```python
#Task-2
#Renaming the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4' using "rename" funtion
df.rename(columns={0: "Random value 1"},inplace=True)
df.rename(columns={1: "Random value 2"},inplace=True)
df.rename(columns={2: "Random value 3"},inplace=True)
df.rename(columns={3: "Random value 4"},inplace=True)
print(df)
```

```python
#Task-3

#Finding the descriptive statistics of the 'df' dataframe
df.describe(include='all')
```

```python
#Task-4

#Check for the null values in 'df' and find the data type of the columns
print(df.isnull())       #checking for null values
print(df.isnull().sum()) #displaying number of null values in each column
```

```python
#Finding the data types of each column
print(df['Random value 1'].dtype)
print(df['Random value 2'].dtype)
print(df['Random value 3'].dtype)
print(df['Random value 4'].dtype)
```

```python
#Task-5

#Display the 'Random value 2' & 'Random value 3' columns with location method
df.loc[0:3,'Random value 2':'Random value 3']
```

```python
#Display the 'Random value 2' & 'Random value 3' columns with index location method 
df.iloc[0:,1:3]
```
